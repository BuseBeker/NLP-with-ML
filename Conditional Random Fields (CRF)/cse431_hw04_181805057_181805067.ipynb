```python
import sklearn_crfsuite
from sklearn_crfsuite import CRF
```

```python
def get_data(txt_file):
  with open(txt_file, encoding = 'utf-8') as f:
    rows = [tuple(row.rstrip('\n').split()) for row in f]
  return rows
```

```python
def lineup_data(datas):
  allOflist=[]
  tempOflist=[]
  for i in datas:
    if len(i)==0:
      allOflist.append(tempOflist)
      tempOflist=[]
      continue
    tempOflist.append(i)
  return allOflist
```

```python
dev_datas = get_data('dev.txt')
test_datas = get_data('test.txt')
train_datas = get_data('train.txt')

train_data = lineup_data(dev_datas)

test_data = lineup_data(test_datas)
```

```python
def word2vec_features(sentence, i):
    word = sentence[i][0]
    features = 
      { 'word': word,
        'isFirst': i == 0,
        'isLast': i == len(sentence) - 1,
        'isCapitalized': word[0].upper() == word[0],
        'isAllCaps': word.upper() == word,
        'isAllLower': word.lower() == word,
        'prefix1': word[0],
        'prefix2': word[:2],
        'prefix3': word[:3],
        'suffix1': word[-1],
        'suffix2': word[-2:],
        'suffix3': word[-3:],
        'prevWord': '' if i == 0 else sentence[i-1][0],
        'nextWord': '' if i == len(sentence)-1 else sentence[i+1][0],
        'hasDigit': any(char.isdigit() for char in word),
        'isAlpha': word.isalpha(), }
    
    return features
```

```python
def extraction_features_data(sentence):
    return [word2vec_features(sentence, i) for i in range(len(sentence))]
```

```python
X_train = [extraction_features_data(sentencee) for sentencee in train_data]
y_train = [ [label for word, label in sentencee] for sentencee in train_data]

X_test = [extraction_features_data(sentencee) for sentencee in test_data]
y_test = [ [label for word, label in sentencee] for sentencee in test_data]

crf = CRF(algorithm='lbfgs', c1=0.1, c2=0.1, max_iterations=100, all_possible_transitions=False)
crf.fit(X_train, y_train)

y_pred = crf.predict(X_test)
```

```python
print(y_pred)
```

```python
list_predicted_tag = []
list_test_tag = []
for i in y_pred:
  for j in i:
    list_predicted_tag.append(j);
for i in y_test:
  for j in i:
    list_test_tag.append(j);
```

```python
from sklearn.metrics import classification_report
print(classification_report(list_test_tag, list_predicted_tag))
```

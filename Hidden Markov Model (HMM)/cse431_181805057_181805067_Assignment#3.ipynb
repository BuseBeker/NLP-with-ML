```python
def get_data(txt_file):
  with open(txt_file, encoding = 'utf-8') as f:
    rows = [tuple(row.rstrip('\n').split()) for row in f]
  return rows
```

```python
dev = get_data('dev.txt')
test = get_data('test.txt')
```

```python
def correct_format_for_data(data):

  list = []
  tempList = []
  for i in data:
    if len(i) == 0:
      list.append(tempList)
      tempList = []
      continue
    tempList.append(i)

  sent_tags = list
  list = []
  for j in sent_tags:
    list.append(j)

  return list
```

```python
data_train = correct_format_for_data(dev)
data_test = correct_format_for_data(test)
```

```python
def correct_format_for_visualize(data):
    
  list=[]
 
  for i in data:
    if len(i)==0:
      continue
    list.append(i)
  return list

visualize_the_train = correct_format_for_visualize(dev)
```

```python
word_tags = {}
for i in data_train:
  for (j,k) in i:
    j = j.lower()
    try:
      try:
        word_tags[k][j]+=1
      except:
        word_tags[k][j] = 1
    except:
      word_tags[k] = {j:1}
    
print(word_tags)
```

```python
#Create an emission probability distribution
emission_probability = {}
for i in word_tags.keys():
  emission_probability[i] = {}
  count = sum(word_tags[i].values())
  for j in word_tags[i].keys():
    emission_probability[i][j] = word_tags[i][j]/count
    
print(emission_probability)
```

```python
tags = {}
for i in visualize_the_train:
  if(i[1] in tags):
    tags[i[1]]+=1
  else:
    tags[i[1]] = 1
    
print(tags)
```

```python
from collections import defaultdict


transitions = defaultdict(lambda: defaultdict(int))

for i in range(1, len(visualize_the_train)):
    prev_tag = visualize_the_train[i-1][1]
    curr_tag = visualize_the_train[i][1]
    transitions[prev_tag][curr_tag] += 1
    
probs = defaultdict(lambda: defaultdict(float))

for prev_tag, next_tag_counts in transitions.items():
    for next_tag, count in next_tag_counts.items():
      probs[prev_tag][next_tag] = count / tags[prev_tag]
    
train_prob =dict(probs)

print(train_prob)
```

```python
def seperate(data):
  wordsOfTest= []
  tagsOfTest = []
  for i in data:
    if len(i)==0:
      continue
    wordsOfTest.append(i[0].lower())
    tagsOfTest.append(i[1])

  return [wordsOfTest, tagsOfTest]

wordsOfTest, tagsOfTest = seperate(test)

print(wordsOfTest)
```

```python
import numpy as np

def viterbi(sentence, tags, probs, emission_prob):
    n = len(sentence)
    viterbi = np.zeros((n, len(tags)))
    backpointer = np.zeros((n, len(tags)))
    for i, word in enumerate(sentence):
        for j, tag in enumerate(tags):
            if i == 0:
                if word in emission_prob[tag]:
                    viterbi[i][j] = emission_prob[tag][word]
                else:
                    viterbi[i][j] = float('-inf')
                backpointer[i][j] = 0
            else:
                max_prob = -1
                max_idx = -1
                for k, prev_tag in enumerate(tags):
                    if word in emission_prob[prev_tag]:
                        prob = viterbi[i-1][k] * probs[prev_tag][tag] * emission_prob[prev_tag][word]
                    else:
                        prob = float('-inf')
                    if prob > max_prob:
                        max_prob = prob
                        max_idx = k
                        viterbi[i][j] = max_prob
                backpointer[i][j] = max_idx
    best_path_prob = max(viterbi[n-1])
    best_path_pointer = np.argmax(viterbi[n-1])
    best_path = [tags[best_path_pointer]]
    for i in range(n-1, 0, -1):
        best_path_pointer = int(backpointer[i][best_path_pointer])
        best_path.append(tags[best_path_pointer])
    return list(reversed(best_path))
```

```python
tagsOfPred=viterbi(wordsOfTest, list(tags), probs, emission_probability)
len(tagsOfPred)
print(tagsOfPred[10:20])
print(tagsOfTest[10:20])
```

```python
from sklearn.metrics import accuracy_score,classification_report

metric= classification_report(tagsOfPred, tagsOfTest)
print(metric)
```
